```python
ticker = "CL=F"
interval = "1d"
start_date = datetime.strptime("2017-01-01", "%Y-%m-%d")
end_date = date.today()

df = yf.download(ticker, start=start_date.strftime("%Y-%m-%d"), end=end_date.strftime("%Y-%m-%d"), interval=interval)


# Plot portfolio spread vs. long run means.
fig = plt.figure()
fig.set_size_inches(16, 6)
plt.plot(df["Adj Close"], label="CL=F")

plt.title("WTI Oil Front Month")
plt.xlabel(f"Time Step ({interval})")
plt.ylabel(f"Price ($/barrel)")
plt.legend()
plt.show()
```
