```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import yfinance as yf
from datetime import date, datetime


sns.set_style("darkgrid")
```

```python
interval = "1d"
end_date = date.today()
start_date = datetime.strptime("2017-01-01", "%Y-%m-%d")
```

```python
# S&P 500 ETF
ticker = "SPY"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
# Micro E-mini S&P 500 Futures
ticker = "MES=F"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
# Micro E-mini S&P 500 Futures
ticker = "MNQ=F"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
# Micro NASDAQ Futures
ticker = "MNQ=F"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
# Micro Dow Futures
ticker = "MYM=F"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
# Micro Russell 2000 Futures
ticker = "M2K=F"
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

print(df.shape)
df.tail()
```

```python
tickers = ["SPY", "MES=F", "MNQ=F", "MYM=F", "M2K=F"]
raw_df = pd.DataFrame(columns=tickers)

for ticker in tickers:
    ticker_df = yf.download(ticker, start=start_date, end=end_date, interval=interval)
    raw_df[ticker] = ticker_df["Adj Close"]
    
print(raw_df.head(2))
print(raw_df.tail(2))
```

```python
# Drop Nans e.g. Micro Futures started after SPY.
df = raw_df.dropna()
df = df - df.mean(axis=0)
df.head(2)
```

```python
df.plot()
```
